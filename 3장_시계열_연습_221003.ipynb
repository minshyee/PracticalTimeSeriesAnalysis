```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = (15, 6)
```

```python
df = pd.read_csv('AirPassengers.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df['#Passengers'].plot()
```

# Moving Average

```python
moving_avg = df['#Passengers'].rolling(window=12).mean()
moving_std = df['#Passengers'].rolling(window=12).std()
expanding_avg = df['#Passengers'].expanding(min_periods=12).mean()


df['#Passengers'].plot()
plt.plot(moving_avg)
plt.plot(moving_std)
plt.plot(expanding_avg)
```
